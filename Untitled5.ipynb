```python
# Import library

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Membaca data

data = pd.read_csv("fifa_world_cup_enhanced_1974_2022.csv")
```

Total Gol per Tahun

```python
#Total Gol per Tahun

goals_year = data.groupby("year")["total_goals"].sum()

plt.figure()
plt.plot(goals_year.index, goals_year.values)
plt.xlabel("Tahun")
plt.ylabel("Total Gol")
plt.title("Total Gol Piala Dunia per Tahun")
plt.show()
```

Total Gol Tim

```python
#Total Gol Tim

home_goals = data.groupby("home_team")["home_goals"].sum()
away_goals = data.groupby("away_team")["away_goals"].sum()
team_goals = home_goals.add(away_goals, fill_value=0)

top_teams = team_goals.sort_values(ascending=False).head(10)

plt.figure()
plt.bar(top_teams.index, top_teams.values)
plt.xticks(rotation=45)
plt.xlabel("Tim")
plt.ylabel("Total Gol")
plt.title("10 Tim dengan Gol Terbanyak")
plt.show()
```

Jumlah Pertandingan per Stage

```python
#Jumlah Pertandingan per Stage

stage_count = data["stage"].value_counts()

plt.figure()
plt.barh(stage_count.index, stage_count.values)
plt.xlabel("Jumlah Pertandingan")
plt.ylabel("Stage")
plt.title("Jumlah Pertandingan berdasarkan Stage")
plt.show()
```

Distribusi Total Gol

```python
#Distribusi Total Gol

plt.figure()
plt.hist(data["total_goals"], bins=10)

plt.xlabel("Jumlah Gol")
plt.ylabel("Frekuensi")
plt.title("Distribusi Total Gol dalam Pertandingan")
plt.show()
```

Scatter Plot Shots vs Goals

```python
#Scatter Plot Shots vs Goals

plt.figure()
plt.scatter(data["shots_home"], data["home_goals"])

plt.xlabel("Shots Home")
plt.ylabel("Home Goals")
plt.title("Hubungan Shots dan Gol Tim Home")
plt.show()
```

Boxplot Total Goals

```python
#Boxplot Total Goals

plt.figure()
plt.boxplot(data["total_goals"])

plt.ylabel("Total Goals")
plt.title("Boxplot Total Gol Pertandingan")
plt.show()
```
